```python
with open ('world_bank_projects.json') as f:
    raw=json.load(f)
```

```python
df_themes = json_normalize(raw, 'mjtheme_namecode',['id'])
df_themes.head(25)
```

```python
df_themes.groupby('id').code.nunique().sort_values(ascending=False).head(25)
```

2.) **Find the top 10 major project themes (using column 'mjtheme_namecode')**

```python
df_themes.groupby('id').code.nunique().sort_values(ascending=False).head(10)
```

<span style="background-color:orange">Secondly, some of the entries have names missings which also skews the data.

<span style="background-color:orange">The task is to not only find a a way to count the major themes but also fill in the missing names that correspond.

```python
df_themes.name.value_counts().head(10)
```

<span style="background-color:orange">I decided to create a dictionary that had all the names as the index to see how many names are in the data set.</span>

```python
df_dict = df_themes.set_index('name').to_dict()
df_dict
```

<span style="background-color:orange">Next I needed to match the names to their corresponding codes. 
However, I needed all the blank names to have corresponding codes. I took the blanks out of the dictionary by using the **.pop()** function.</span>

```python
df_dict['code']
```

```python
df_dict['code'].pop('')
df_dict
```

<span style="background-color:orange"> Once I had the names corresponding to the codes correctly, I needed to flip the keys/values so that my "fill from dictionary" trick I just learned would work. </span>

```python
code_dict = dict(zip(df_dict['code'].values(),df_dict['code'].keys()))
code_dict
```

```python
#Empty DataFrame
code_df = df_themes.set_index('code')                            
code_df.head(25)
```

3.) **Create a dataframe with the missing names filled in.**

```python
#Updated DataFrame
code_df['name'].update(pd.Series(code_dict))
code_df.head(100)
```
